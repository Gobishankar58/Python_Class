```python
import cProfile

def slow_function():
    total = 0
    for i in range(1000000):
        total += i
    return total

def fast_function():
    return sum(range(1000000))

def profile_function():
    cProfile.run('slow_function()')

if __name__ == '__main__':
    # Profile the slow_function
    profile_function()

    # Optimize the slow_function using a faster implementation
    result = fast_function()
    print("Fast function result:", result)
```
