#  Context managers and the `with` statement

```python
file = open("hello.txt", "w")
file.write("Hello, World!")
file.close()
```

```python
file = open("hello.txt", "w")
```

```python
try:
    file.write("Hello, World!")
except Exception as e:
    print(f"An error occurred while writing to the file: {e}")
finally:
    file.close()
```

```python
with open("hello.txt", mode="w") as file:
    file.write("Hello, World")
```

```python
class File_context:
    def __init__(self,name,mode):
        self.file=None
        self.name=name
        self.mode=mode
    
    def __enter__(self):
        self.file=open(self.name,self.mode)
        self.file.write("Thsi is start")
        return self.file
    
    def __exit__(self,*args):
        self.file.write("This was written by exit")
        self.file.close()
        
with File_context('mytext.txt','w') as myfile:
    myfile.write("My data")
    
```
